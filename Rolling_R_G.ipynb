## Examining Rolling Runs Scored per Game 

```python
%pip install pybaseball -q
import pybaseball as pyb
import pandas as pd
import numpy as np
from pybaseball import team_game_logs
```

```python
teams = ["BAL", "TBR", "BOS", "NYY", "TOR", "MIN", "DET", "CHW", "KCR", "CLE",
         "LAA", "OAK", "TEX", "HOU", "SEA", "LAD", "SFG", "ARI", "COL", "SDP",
         "MIL", "STL", "CHC", "CIN", "PIT", "NYM", "WSN", "ATL", "PHI", "MIA"]
```

```python
bat_BAL_18 = team_game_logs(2021, "BAL", "batting")
scored = bat_BAL_18['R']
```

```python
runs = sum(scored)
runs_per_G = round(runs / 162, 1)
runs_per_G
```

```python
df = pd.DataFrame()

for team in teams:
  bat_21 = team_game_logs(2021, team, "batting")
  scored = bat_21['R']

  rolling_avg = []
  upper_bound = 14

  for i in range(len(scored) + 1):
    if upper_bound <= 161:
      avg = round(sum(scored[i:upper_bound]) / 15, 1)
      
      rolling_avg.append(avg)
      
      upper_bound += 1

  df[team] = rolling_avg
```

```python
df
```

```python
runs_per_game = pd.DataFrame()

for team in teams: 
  r_g = []
  bat_logs = team_game_logs(2021, team, "batting")
  runs_scored = bat_logs['R']

  average = round(sum(runs_scored) / len(runs_scored), 1)
  r_g.append(average)

  runs_per_game[team] = r_g
```
