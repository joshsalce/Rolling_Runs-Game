```python
%pip install pybaseball -q
import pybaseball as pyb
import pandas as pd
import numpy as np
from pybaseball import team_game_logs
```

```python
teams = ["BAL", "TBR", "BOS", "NYY", "TOR", "MIN", "DET", "CHW", "KCR", "CLE",
         "LAA", "OAK", "TEX", "HOU", "SEA", "LAD", "SFG", "ARI", "COL", "SDP",
         "MIL", "STL", "CHC", "CIN", "PIT", "NYM", "WSN", "ATL", "PHI", "MIA"]
```

```python
df = pd.DataFrame()

for team in teams:
  pitching_21 = team_game_logs(2021, team, "pitching")
  allowed = pitching_21['R']

  rolling_avg = []
  upper_bound = 14

  for i in range(len(allowed) + 1):
    if upper_bound <= 161:
      avg = round(sum(allowed[i:upper_bound]) / 15, 1)
      
      rolling_avg.append(avg)
      
      upper_bound += 1

  df[team] = rolling_avg
```

```python
runs_allowed_per_game = pd.DataFrame()

for team in teams: 
  ra_g = []
  pitch_logs = team_game_logs(2021, team, "pitching")
  runs_allowed = pitch_logs['R']

  average = round(sum(runs_allowed) / len(runs_allowed), 1)
  ra_g.append(average)

  runs_allowed_per_game[team] = ra_g
```
